```python
#What seasons did the study focus on?
#KEY for 3.7.1c, 3.7.2c and 3.7.3c
#1: winter season
#2: summer season
#3: Other season
#This gives an indication of the messiness of the data
CEEPA_df['seas3nam'].value_counts()
```

```python
#What was the breakdown of households by country?
CEEPA_df['country'].value_counts()
```

```python
#What sizes are the farms surveyed? (KEY  1: small-scale; 2: medium scale; 3: large-scale)
#More than half of the farms surveyed are self-reported as small-scale farms
CEEPA_df['farmtype'].value_counts(normalize=True)
```

```python
#When we look at country, do they tend to have a farm type?
#In Zimbabwe, Zambia, Ghana, Ethiopia, and Egypt most farms are self-reported as small scale
#No pattern in regions
#In Senegal, most farms are reported as large-scale
#In Burkinafaso, most farms are small and medium scale
CEEPA_df.groupby('country')['farmtype'].value_counts(normalize=True).sort_index(ascending=True)
```

```python
#Do most farms have electricity? (key 1 = YES, 2 = NO)
#59% households DO NOT have electricity 
#31% household have electricity 
CEEPA_df['electricity'].value_counts(normalize=True)
```

```python
#USE OPEN REFINE TO GROUP THESE INTO SOUTH, EAST, WEST, NORTH AFRICA???
#59% households DO NOT have electricity 
#31% household have electricity 
#Are certain countries less likely to have electricity?
#No pattern by region
#99% households surveyed in Burkinafaso don't have electricity !?!
#Zimbabwe, Zambia, Senegal, and Kenya have > 80% households without electricity
CEEPA_df.groupby('country')['electricity'].value_counts(normalize=True).sort_values(ascending=False).plot(kind='bar')
```

```python
#Do farms without electricity tend to be small scale? 
#electricity key (1 = YES, 2 = NO)
#Nope, there is no pattern between size of farm and electricity
CEEPA_df.groupby('farmtype')['electricity'].value_counts(normalize=True)
```

```python
#How dependent are these households on their income from farming?
#What percentage of total net house income comes from non-farming?
#In 75% of households, the percentage of income that comes non-farming is about 30%
#This means in 75% of households, the percentage of income that comes from farming is 70%
#So climate changes and adaptations are very relevent for these households
CEEPA_df.boxplot(column='income_nonfarm', showfliers=False)
```

```python
#Number of years of farming experience
#Obvi 790 and 750 years are typos 
CEEPA_df['farmingexperience'].dropna().value_counts().sort_index(ascending=False)
```

```python
#The median years of farming experience is around 20 and 75th percentile is 30 years
CEEPA_df['farmingexperience'].describe()
```

```python
#Removing outliers, we can clearly see the distribution of farming years more clearly
CEEPA_df.boxplot(column='farmingexperience', showfliers=False)
```

```python
#A histogram of farming experience again shows that most farmers have about 0-25 years of farming experience
CEEPA_df['farmingexperience'].dropna().value_counts().plot(kind='hist',figsize=(20,20))
```

```python
#What are the most frequently reported religious beliefs?
#KEY for religion: 1. Nonreligious; 2: Christianity; 3: Islam; 4; African Traditional Religion; 5: Other 
#Most households identify as Christian or Islamic
CEEPA_df['religion'].value_counts(normalize=True).dropna()
```

```python
#How far is the nearest market to sell harvest?
#CEEPA_df.boxplot(column='distsmktkm',showfliers=False)
#Not too far -- 75% of households travel for 1 hour or less to get to the market
CEEPA_df['distsmkthr'].describe()
```

```python
#How do households typically transport their harvest?
#KEY = 1. walk; 2.animal; 3. cart/bicycle; 4. truck or other motorized vehicle; 6 . Other
#The highest number of people either take a motor vehicle (truck/car) or they walk ... extremes
CEEPA_df['transport'].value_counts().plot(kind='bar',figsize=(10,10))
```

```python
#Does transportation differ by country?
#In Egypt and South Africa, the most frequent form of transportation is motor vehicle
#In Niger, the most frequent form of transportation to get to market is walking 
CEEPA_df.groupby('country')['transport'].value_counts(normalize=True)
```

```python
#Does transportation differ by farm type?
#Not surprisingly, larger farms transport harvest mainly by motor vehicle or cart
CEEPA_df.groupby('farmtype')['transport'].value_counts(normalize=True)
```

```python
#Do extension officers provide information about expected rainfall and precipitation? 
#KEY 1: yes; 2: no
#75% households say they don't receive information from extension officers
CEEPA_df['extclim'].value_counts(normalize=True)
```
