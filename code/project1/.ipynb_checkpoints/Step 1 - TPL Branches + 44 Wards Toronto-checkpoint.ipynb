# Importing Libraries

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
% matplotlib inline
import numpy as np
```

# Creating Two Geo Data Frames

```python
# Geo df for the 44 wards of Toronto
# sourced from - https://www1.toronto.ca/wps/portal/contentonly?vgnextoid=b1533f0aacaaa210VgnVCM1000006cd60f89RCRD
wards = gpd.read_file('wards_may2010/tcl3_icitw.shp')
```

```python
wards.head()
```

```python
#Geo df for branches of the library
df = pd.read_excel('eProduct_Symphony_BranchDetails_201706/BranchBridgeTable.xls')
points = df.apply (lambda row: Point(row.LONGITUDE, row.LATITUDE), axis = 1)
branches = gpd.GeoDataFrame(df, geometry = points)
branches.crs = {'init': 'epsg:4267'}
branches
```

# Assigning Projections for each Geo Data Frame 

```python
#The city of Toronto used NAD27 as their datum, we should likely use the same for the branches (epsg = 4267)
wards.crs
```

```python
ax = wards.plot(figsize=(10,10),color = 'grey', edgecolor = 'white', linewidth = 2)
ax.axis('off')
```

```python
ax = branches.plot(figsize=(10,10), color='#E81B49', markersize = 5, alpha= 0.5)
ax.axis('off')
```

```python
# Set the CRS to be the same for branches and wards
wards.to_crs(branches.crs, inplace=True)
```

```python
branches.crs
```

```python
wards.crs
```

# Plotting TPL branches over Toronto's wards

```python
ax = wards.plot(figsize=(20,20),color = 'grey', edgecolor = 'white', linewidth = 2)
ax = branches.plot(figsize=(20,20), color='#E81B49', markersize = 6, alpha= 0.5, ax=ax)
ax.axis('off')
```

# Spatial Join - which ward is each branch in?

```python
branches.head()
```

```python
#76 branches in our data set
branches.shape
```

```python
#Cleaning the NaN values 
branches.dropna(inplace=True)
pd.set_option("display.max_columns", 100)
branches.head(2)
```

```python
#We want to keep every single row in the branches data set even if it isn't in a ward
branches_within_wards = gpd.sjoin(branches.reset_index(), wards, how='left', op='within')
branches_within_wards.head()
```

```python
#Still 76 branches within wards
branches_within_wards.shape
```

```python
#Sent to library to confirm ward assignments for a few branches (didn't match the ward assignment we get from the sjoin)
df.to_csv("branches_within_wards.csv", index=False)
```

# Merging geo data frame with ebook circ dataset

```python
ebooks = pd.read_excel('eProduct_Symphony_BranchDetails_201706/eProduct_CirculationActivityDetail-201706.xlsx')
ebooks.dtypes
```

```python
ebooks.head(3)
```

```python
#Checking to see if the "Branch" column from ebooks df matches "OverDrive Branch" column from the branches_within_wards df
ebooks[ebooks['Branch'].str.contains("Annette", na=False)]
```

```python
branches_ebooks = branches_within_wards.merge(ebooks, left_on="OverDrive Branch", right_on="Branch")
branches_ebooks.head()
```

```python
#Ebook circulation by branches
branches_ebooks.Branch.value_counts()
```

```python
#Ebook circulation by wards
branches_ebooks.NAME.value_counts()
```
